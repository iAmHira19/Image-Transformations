```python
# Step 1: Install OpenCV
!pip install opencv-python-headless

# Step 2: Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images using Matplotlib
def display_image(img, title="Image"):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Step 3: Load the image (upload an image or use an online URL)
from google.colab import files
uploaded = files.upload()

# Load the image
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)

# Step 4: Translation Transformation
def translate_image(img, x_shift, y_shift):
    rows, cols = img.shape[:2]
    translation_matrix = np.float32([[1, 0, x_shift], [0, 1, y_shift]])
    return cv2.warpAffine(img, translation_matrix, (cols, rows))

# Step 5: Scaling Transformation
def scale_image(img, scale_x, scale_y):
    return cv2.resize(img, None, fx=scale_x, fy=scale_y, interpolation=cv2.INTER_LINEAR)

# Step 6: Rotation Transformation
def rotate_image(img, angle):
    rows, cols = img.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    return cv2.warpAffine(img, rotation_matrix, (cols, rows))

# Step 7: Apply the transformations
translated_image = translate_image(image, 100, 50)  # Translate image
scaled_image = scale_image(image, 0.5, 0.5)         # Scale image (reduce size by half)
rotated_image = rotate_image(image, 45)             # Rotate image by 45 degrees

# Step 8: Display the original and transformed images
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
display_image(image, "Original Image")

plt.subplot(2, 2, 2)
display_image(translated_image, "Translated Image (100, 50)")

plt.subplot(2, 2, 3)
display_image(scaled_image, "Scaled Image (50%)")

plt.subplot(2, 2, 4)
display_image(rotated_image, "Rotated Image (45 degrees)")

# Step 9: Save the transformed images (optional)
cv2.imwrite('translated_image.jpg', translated_image)
cv2.imwrite('scaled_image.jpg', scaled_image)
cv2.imwrite('rotated_image.jpg', rotated_image)
```
